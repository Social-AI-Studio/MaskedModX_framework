# Computation of BLEURT scores and Wilcoxon signed-rank tests

```python
def get_BLEURT_scores_and_perform_Wilcoxon_test(df, scorer):
    references = df["baseline_response_text"].tolist()
    BLEURT_scores = {}
    wilcoxon_results_data = []
    for col in df.columns:
        if "response_text" in col and col != "baseline_response_text":
            candidates = df[col].tolist()
            scores = scorer.score(references=references, candidates=candidates)
            BLEURT_scores[col] = scores
            mean_score = np.mean(scores)
            sd = np.std(scores, ddof=1)
            print(col)
            print("mean_BLEURT")
            print(mean_score)
            print("sd_BLEURT")
            print(sd)
    instructions = ["persona_instruction1", "persona_instruction2", "persona_instruction3"]
    for instruction in instructions:
        significant_pairs = 0
        personas_groupby_instruction = sorted([persona for persona in BLEURT_scores.keys() if instruction in persona])
        pairs = list(itertools.combinations(personas_groupby_instruction, 2))
        print("pairs")
        print(pairs)
        for pair in pairs:
            test = pg.wilcoxon(x=BLEURT_scores[pair[1]], y=BLEURT_scores[pair[0]], alternative="two-sided", correction=False)
            print(pair)
            print("Wilcoxon signed-rank test")
            print(test)
            p_value = test["p-val"].iloc[0]
            if p_value < .05:
                significant_pairs += 1
            wilcoxon_results_data.append({
                "instruction_group": f"{instruction}_response_text",
                "persona1": pair[0],
                "persona2": pair[1],
                "p_value": p_value,
                "RBC": test["RBC"].iloc[0]
            })
        print(instruction)
        print("RIS")
        print(significant_pairs/len(pairs))
    return pd.DataFrame(wilcoxon_results_data)
```

```python
BLEURT_HatReD_reasonings_gender_identity_personas_llama4_scout_instruct_basic = get_BLEURT_scores_and_perform_Wilcoxon_test(df=HatReD_reasonings_gender_identity_personas_llama4_scout_instruct_basic, scorer=scorer)
print("----------------------------------------------------------------------")
BLEURT_HatReD_reasonings_race_ethnicity_personas_llama4_scout_instruct_basic = get_BLEURT_scores_and_perform_Wilcoxon_test(df=HatReD_reasonings_race_ethnicity_personas_llama4_scout_instruct_basic, scorer=scorer)
print("----------------------------------------------------------------------")
BLEURT_HatReD_reasonings_religion_personas_llama4_scout_instruct_basic = get_BLEURT_scores_and_perform_Wilcoxon_test(df=HatReD_reasonings_religion_personas_llama4_scout_instruct_basic, scorer=scorer)
print("----------------------------------------------------------------------")
BLEURT_HatReD_reasonings_sexual_orientation_personas_llama4_scout_instruct_basic = get_BLEURT_scores_and_perform_Wilcoxon_test(df=HatReD_reasonings_sexual_orientation_personas_llama4_scout_instruct_basic, scorer=scorer)
print("----------------------------------------------------------------------")
```

```python
%%R
persona_map <- c(
    "a_man" = "a man",
    "a_woman" = "a woman",
    "a_transgender_man" = "a transgender man",
    "a_transgender_woman" = "a transgender woman",
    "a_non_binary_individual" = "a non-binary individual",
    "an_Asian_Asian_American_individual" = "an Asian/Asian American individual",
    "a_Black_African_American_individual" = "a Black/African American individual",
    "a_Hispanic_Latinx_individual" = "a Hispanic/Latinx individual",
    "a_White_Caucasian_individual" = "a White/Caucasian individual",
    "a_Native_Hawaiian_Pacific_Islander_individual" = "a Native Hawaiian/Pacific Islander individual",
    "a_Native_American_First_Nations_individual" = "a Native American/First Nations individual",
    "a_heterosexual_straight_individual" = "a heterosexual/straight individual",
    "a_gay_lesbian_individual" = "a gay/lesbian individual",
    "a_bisexual_individual" = "a bisexual individual",
    "a_Muslim" = "a Muslim",
    "an_atheist" = "an atheist",
    "a_Buddhist" = "a Buddhist",
    "a_Christian" = "a Christian",
    "a_Hindu" = "a Hindu",
    "a_Jew" = "a Jew"
)
generate_plot <- function(df, outputfile) {
    metadata <- list(
        list(name = "persona_instruction1",
             pattern = "persona_instruction1_response_text",
             subplot_title = "(a) Instruction 1"),
        list(name = "persona_instruction2",
             pattern = "persona_instruction2_response_text",
             subplot_title = "(b) Instruction 2"),
        list(name = "persona_instruction3",
             pattern = "persona_instruction3_response_text",
             subplot_title = "(c) Instruction 3")
    )
    plot_list <- list()
    for (grp in metadata) {
        instruction_subset <- df[df$instruction_group == grp$pattern, ]
        instruction_subset$persona1_label <- unname(persona_map[sub("_in_the_United_States", "", sub(paste0("_", grp$pattern), "", instruction_subset$persona1, fixed = TRUE), fixed = TRUE)])
        instruction_subset$persona2_label <- unname(persona_map[sub("_in_the_United_States", "", sub(paste0("_", grp$pattern), "", instruction_subset$persona2, fixed = TRUE), fixed = TRUE)])
        instruction_subset$persona1_factor <- factor(instruction_subset$persona1_label, levels = sort(unique(c(instruction_subset$persona1_label, instruction_subset$persona2_label))))
        instruction_subset$persona2_factor <- factor(instruction_subset$persona2_label, levels = sort(unique(c(instruction_subset$persona1_label, instruction_subset$persona2_label))))
        instruction_subset$legend_label <- ifelse(
            instruction_subset$p_value < 0.05 & instruction_subset$RBC < 0,
            "Baseline NLEs are significantly more semantically similar to x-axis persona NLEs",
            ifelse(
                instruction_subset$p_value < 0.05 & instruction_subset$RBC > 0,
                "Baseline NLEs are significantly more semantically similar to y-axis persona NLEs",
                "nonsignificant"
            )
        )
        g <- ggplot(instruction_subset, aes(persona1_factor, persona2_factor, fill = legend_label)) +
            geom_tile(color = "white", linewidth = 1, show.legend = TRUE) +
            scale_fill_manual(
                values = c(
                    "Baseline NLEs are significantly more semantically similar to x-axis persona NLEs" = "#840032",
                    "Baseline NLEs are significantly more semantically similar to y-axis persona NLEs" = "#e59500",
                    "nonsignificant" = "#002642"
                ),
                limits = c(
                    "Baseline NLEs are significantly more semantically similar to x-axis persona NLEs",
                    "Baseline NLEs are significantly more semantically similar to y-axis persona NLEs",
                    "nonsignificant"
                ),
                name = "Wilcoxon Signed-Rank Test Results",
                drop = FALSE
            ) +
            labs(
                title = grp$subplot_title,
                x = NULL,
                y = NULL
            ) +
            coord_fixed() +
            theme_minimal(base_family = "Helvetica", base_size = 42) +
            theme(
                axis.text.x = element_text(angle = 30, hjust = 1),
                plot.title = element_text(hjust = 0.5, margin = margin(b = 3, unit = "pt")),
                panel.grid.major = element_line(linewidth = 0.5, color = "grey80")
            )
        if (grp$name == "persona_instruction1") {
            g <- g + theme(plot.margin = margin(t = 0, r = 25, b = 0, l = 0, unit = "pt"))
        } else if (grp$name == "persona_instruction3") {
            g <- g + theme(
                axis.title.y = element_blank(),
                axis.text.y = element_blank(),
                plot.margin = margin(t = 0, r = 0, b = 0, l = 25, unit = "pt")
            )
        } else {
            g <- g + theme(
                axis.title.y = element_blank(),
                axis.text.y = element_blank(),
                plot.margin = margin(t = 0, r = 25, b = 0, l = 25, unit = "pt")
            )
        }
        plot_list[[grp$name]] <- g
    }
    aggregated_plot <- wrap_plots(plot_list, nrow = 1, guides = "collect") & theme(
        text = element_text(family = "Helvetica", size = 42),
        legend.title = element_text(margin = margin(b = 3, unit = "pt")),
        legend.position = "bottom",
        legend.justification = 0,
        legend.direction = "vertical"
    )
    ggsave(
        outputfile,
        plot = aggregated_plot,
        device = cairo_ps,
        bg = "white",
        width = 28,
        height = 14,
        units = "in"
    )
}
```
